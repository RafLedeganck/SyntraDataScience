## Required packages

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

## Image file

```python
im_source = np.array(Image.open('lightbulb_small.jpg'))
plt.imshow(im_source)
plt.show()
```

## Exercise 1: Repeat

```python
repeat3x8 = np.tile(im_source, (3, 8, 1))
plt.imshow(repeat3x8)
plt.show()
```

## Exercise 2: Repeat, flip & reverse

```python
repeat_x8 = np.tile(im_source, (1, 8, 1))
repeat_fvx8 = np.tile(np.flip(im_source,1), (1, 8, 1))
repeat_fhx8 = np.tile(np.flip(im_source,0), (1, 8, 1))
repeat_fhvx8 = np.tile(np.flip(np.flip(im_source,1),0), (1, 8, 1))

im_2 = np.append(np.append(repeat_x8, repeat_fvx8, axis=0),np.append(repeat_fhx8, repeat_fhvx8, axis=0), axis=0)

plt.imshow(im_2)
plt.show()
```

## Exercise 3: Playing with color

```python
im_blue = im_source.copy()
im_blue[:,:,2] = 255

im_green = im_source.copy()
im_green[:,:,1] = 255

im_fuchsia = im_source.copy()
im_fuchsia[:,:,[0,2]] = 255

im_large = im_source.repeat(2, axis = 0).repeat(2, axis = 1)

im_top = np.tile(im_blue, (1,4,1))
im_mid = np.append(np.append(np.tile(im_green, (2, 1, 1)),im_large,axis=1), np.tile(im_green, (2,1,1)), axis=1)
im_bottom = np.tile(im_fuchsia, (1,4,1))

im_3 = np.append(np.append(im_top, im_mid, axis=0), im_bottom, axis=0)
plt.imshow(im_3)
plt.show()     
```

## Exercise 4: Negative color

```python
im_neg = 255 - im_source
plt.imshow(im_neg)
plt.show()   
```

## Exercise 5: Resize & Overlay

```python
im_small = np.array(Image.open('lightbulb_small.jpg').resize((100,100)))
im_overlay = im_source.copy()
im_overlay[175:255, 255:315] = im_small[10:90,20:80]
plt.imshow(im_overlay)
plt.show()   
```
